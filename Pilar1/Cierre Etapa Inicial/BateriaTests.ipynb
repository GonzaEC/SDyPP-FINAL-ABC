```python
%%bash
for test in "hola mundo|0" "hola mundo|00" "hola mundo|000" "hola mundo|0000" "hola mundo|00000" "hola mundo|000000" "blockchain|0" "blockchain|0000" "blockchain|00000"; do
    base=$(echo $test | cut -d'|' -f1)
    prefix=$(echo $test | cut -d'|' -f2)
    ./brute_force_timed "$base" "$prefix"
    echo "---"
done
```

```python
import hashlib
import time

tests = [
    ("hola mundo", "0"),
    ("hola mundo", "00"),
    ("hola mundo", "000"),
    ("hola mundo", "0000"),
    ("hola mundo", "00000"),
    ("hola mundo", "000000"),
    ("blockchain", "0"),
    ("blockchain", "0000"),
    ("blockchain", "00000"),
]

for base, prefix in tests:
    start = time.time()
    nonce = 0
    while True:
        text = base + str(nonce)
        h = hashlib.md5(text.encode()).hexdigest()
        if h.startswith(prefix):
            elapsed = (time.time() - start) * 1000
            print(f"Base: {base} | Prefijo: {prefix} | Nonce: {nonce} | Hash: {h} | Tiempo: {elapsed:.3f} ms")
            break
        nonce += 1
```
