```python
%%writefile hello.cu
#include <stdio.h>

__global__ void helloFromGPU() {
    printf("Hola mundo desde la GPU! Bloque %d, Hilo %d\n", blockIdx.x, threadIdx.x);
}

int main() {
    printf("Hola mundo desde la CPU!\n");

    helloFromGPU<<<2, 4>>>();

    cudaDeviceSynchronize();

    return 0;
}
```
