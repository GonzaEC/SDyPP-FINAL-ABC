```python
import hashlib

def brute_force_cpu(base, prefix):
    nonce = 0
    while True:
        text = base + str(nonce)
        hash_result = hashlib.md5(text.encode()).hexdigest()
        if hash_result.startswith(prefix):
            print(f"Nonce encontrado: {nonce}")
            print(f"Hash resultante:  {hash_result}")
            print(f"Input completo:   {text}")
            return
        nonce += 1

brute_force_cpu("hola mundo", "0000")
```
